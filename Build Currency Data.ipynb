```python
#import packages
import pandas as pd

#read the file
df = pd.read_csv('currency.csv', header=1)

indian_key = 'EXR.D.INR.EUR.SP00.A'
sterling = 'EXR.D.GBP.EUR.SP00.A'
us_dollar = 'EXR.D.USD.EUR.SP00.A'
canadian = 'EXR.D.CAD.EUR.SP00.A'
curr_key = [indian_key, sterling, us_dollar, canadian]
df.drop([0, 1, 2, 3, 4, 5], axis=0, inplace=True)
#print the head
df.head(10)
```

```python
df.info()
```

```python
names = df.columns
date_name = 'Unnamed: 0'
curr = []
for name in names:
    if 'EXR.D' not in name and name != date_name:
        df.drop(name, axis=1, inplace=True)
    else:
        curr.append(name)
df.columns = ['date'] + curr[1:]
df.head()
```
